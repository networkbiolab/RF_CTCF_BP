```python
#SK-N-SH
a = pd.read_csv("./SK/Motif.txt", sep="\s+")
b = pd.read_csv("./SK/ATAC-seq.txt", sep="\s+")

c = pd.read_csv("./SK/H3K9ac.txt", sep="\s+")
d = pd.read_csv("./SK/H3K27ac.txt", sep="\s+")
e = pd.read_csv("./SK/H3K4me3.txt", sep="\s+")
f = pd.read_csv("./SK/H3K4me2.txt", sep="\s+")
g = pd.read_csv("./SK/H3K4me1.txt", sep="\s+")
h = pd.read_csv("./SK/H3K79me2.txt", sep="\s+")
i = pd.read_csv("./SK/H3K9me3.txt", sep="\s+")
j = pd.read_csv("./SK/H3K27me3.txt", sep="\s+")
SK = pd.read_csv("./SK/H3K36me3.txt", sep="\s+")
l = pd.read_csv("./SK/H4K20me1.txt", sep="\s+")
SK = pd.read_csv("./SK/H2AFZ.txt", sep="\s+")

n = pd.read_csv("./SK/Methyl.txt", sep="\s+")

values = pd.read_csv("./SK/values.txt", sep="\s+")

df = a

for x in [b,c,d,e,f,g,h,i,j,k,l,m,n,values]:
        df=pd.merge(df,x,on="pos")

SK = df.drop(['value','pos'], axis=1)
SK_values = df['value']
print(SK.head)
print(SK_values.head)
```

```python
# Concats

GK = pd.concat([G,K])
GK_values = pd.concat([G_values,K_values])

GKM = pd.concat([GK,HL])
GKM_values = pd.concat([GK_values,HL_values])

GKSK = pd.concat([GK,SK])
GKSK_values = pd.concat([GK_values,SK_values])

GM = pd.concat([G,HL])
GM_values = pd.concat([G_values,HL_values])

GMSK = pd.concat([GM,SK])
GMSK_values = pd.concat([GM_values,SK_values])

KM = pd.concat([K,HL])
KM_values = pd.concat([K_values,HL_values])

KMSK = pd.concat([KM,SK])
KMSK_values = pd.concat([KM_values,SK_values])
```

```python
# Todo contra GM

X_train = KMSK
y_train = KMSK_values
X_test = G
y_test = G_values

from sklearn.ensemble import RandomForestClassifier
# random forest model creation
rfc = RandomForestClassifier(n_estimators=3000,min_samples_split=20, 
                             min_samples_leaf=10,max_features='auto',max_depth=None)
rfc.fit(X_train,y_train)
# predictions
y_pred = rfc.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rfc.classes_)
disp.plot()
```
